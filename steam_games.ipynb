Extracion del archivo USER_ITEMS.JSON

```python
# Este código carga datos desde un archivo JSON ('user_items.json') en un DataFrame de pandas ('df_user_item').
# Utiliza ast.literal_eval() para convertir cada línea del archivo JSON en un diccionario de Python,
# luego crea el DataFrame con estos diccionarios y muestra una muestra aleatoria de 10 filas.

rows = []

with open('./user_items.json', encoding='utf-8') as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))

df_user_item = pd.DataFrame(rows)
df_user_item.sample(10)       
```

```python
# Este código utiliza el método isna() de DataFrame para identificar valores faltantes (NaN) en cada columna,
# y luego suma la cantidad de valores faltantes por columna. 
df_user_item.isna().sum() 
```

```python
# El código 'df_user_item.shape' devuelve una tupla que representa las dimensiones del DataFrame df_user_item.
# El primer elemento de la tupla indica el número de filas en el DataFrame, mientras que el segundo elemento indica el número de columnas.
df_user_item.shape
```

```python
# devuelve una serie booleana que indica si cada fila en el DataFrame df_user_item es una duplicada de una fila anterior.
# True indica que la fila es una duplicada, mientras que False indica que no lo es.
##df_user_item.duplicated()
```

```python
# Este código desanida los datos anidados dentro del DataFrame df_user_item, donde cada fila contiene información de un usuario, incluyendo una lista de 'items' que representa los juegos que posee.
# Itera sobre cada fila del DataFrame para extraer información relevante del usuario, como 'user_id', 'steam_id' y 'user_url'.
# Luego, para cada fila, itera sobre la lista de 'items', extrayendo los datos de cada juego ('item_id', 'item_name', 'playtime_forever', 'playtime_2weeks') y crea nuevas filas con esta información desanidada.
# Estas nuevas filas se agregan a la lista 'datos_desanidados'.
# Finalmente, se utiliza esta lista para crear un nuevo DataFrame df_items_desanidado, el cual contiene los datos desanidados de los juegos para facilitar su análisis posterior.

datos_desanidados = []


for index, row in df_user_item.iterrows():
    user_id = row['user_id']
    steam_id = row['steam_id']
    user_url = row['user_url']
    
    for item in row['items']:
        new_row = {
            'user_id': user_id,
            'steam_id': steam_id,
            'user_url': user_url,
            'item_id': item['item_id'],
            'item_name': item['item_name'],
            'playtime_forever': item['playtime_forever'],
            'playtime_2weeks': item['playtime_2weeks']
        }
        
        datos_desanidados.append(new_row)

df_items_desanidado = pd.DataFrame(datos_desanidados)

df_items_desanidado
```
